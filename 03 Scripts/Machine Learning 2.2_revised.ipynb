```python
def summarize_station_coverage(df_weather_filtered, separator='_'):
    """
    Prints a compact summary of which observation types exist per station.
    """
    # Split column names
    split_cols = [col.split(separator, 1) for col in df_weather_filtered.columns if separator in col]
    df_split = pd.DataFrame(split_cols, columns=['station', 'observation'])

    all_obs = set(df_split['observation'])
    summary = {}

    for station, group in df_split.groupby('station'):
        present = set(group['observation'])
        missing = sorted(all_obs - present)
        summary[station] = {
            'count': len(present),
            'missing': missing
        }

    # Pretty print
    for station, info in summary.items():
        print(f"{station}: {info['count']} types", end='')
        if info['missing']:
            print(f"  (missing: {', '.join(info['missing'])})")
        else:
            print()
```

```python
"""
Data Cleaning Utility Functions
A collection of common data cleaning operations for pandas DataFrames
"""

import pandas as pd
import numpy as np
import re
from typing import List, Union, Optional


# ============================================================================
# === UTILITY FUNCTIONS (scroll past to see example usage) ==================
# ============================================================================


def clean_column_names(df: pd.DataFrame) -> pd.DataFrame:
    """
    Standardize column names: lowercase, replace spaces with underscores,
    remove special characters.
    """
    df = df.copy()
    df.columns = (
        df.columns
        .str.lower()
        .str.replace(' ', '_')
        .str.replace('[^a-z0-9_]', '', regex=True)
    )
    return df


def remove_empty_columns(df: pd.DataFrame, threshold: float = 1.0) -> pd.DataFrame:
    """
    Remove columns where percentage of missing values exceeds threshold.
    
    Args:
        df: Input DataFrame
        threshold: Proportion of missing values (0-1). Default 1.0 removes only fully empty columns
    """
    df = df.copy()
    missing_pct = df.isnull().sum() / len(df)
    cols_to_keep = missing_pct[missing_pct < threshold].index
    return df[cols_to_keep]


def remove_empty_rows(df: pd.DataFrame, threshold: float = 1.0) -> pd.DataFrame:
    """
    Remove rows where percentage of missing values exceeds threshold.
    
    Args:
        df: Input DataFrame
        threshold: Proportion of missing values (0-1). Default 1.0 removes only fully empty rows
    """
    df = df.copy()
    missing_pct = df.isnull().sum(axis=1) / len(df.columns)
    return df[missing_pct < threshold]


def detect_outliers_iqr(df: pd.DataFrame, columns: Optional[List[str]] = None, 
                        multiplier: float = 1.5) -> pd.DataFrame:
    """
    Detect outliers using IQR method for specified numeric columns.
    Returns a boolean DataFrame indicating outliers.
    
    Args:
        df: Input DataFrame
        columns: List of columns to check. If None, checks all numeric columns
        multiplier: IQR multiplier (1.5 is standard, 3.0 is more conservative)
    """
    if columns is None:
        columns = df.select_dtypes(include=[np.number]).columns.tolist()
    
    outliers = pd.DataFrame(index=df.index)
    
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - multiplier * IQR
        upper_bound = Q3 + multiplier * IQR
        outliers[col] = (df[col] < lower_bound) | (df[col] > upper_bound)
    
    return outliers


def handle_outliers(df: pd.DataFrame, columns: Optional[List[str]] = None, 
                   method: str = 'cap', multiplier: float = 1.5) -> pd.DataFrame:
    """
    Handle outliers by capping, removing, or replacing with NaN.
    
    Args:
        df: Input DataFrame
        columns: Columns to process. If None, processes all numeric columns
        method: 'cap' (winsorize), 'remove', or 'nan'
        multiplier: IQR multiplier
    """
    df = df.copy()
    
    if columns is None:
        columns = df.select_dtypes(include=[np.number]).columns.tolist()
    
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - multiplier * IQR
        upper_bound = Q3 + multiplier * IQR
        
        if method == 'cap':
            df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)
        elif method == 'remove':
            df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
        elif method == 'nan':
            df.loc[(df[col] < lower_bound) | (df[col] > upper_bound), col] = np.nan
    
    return df


def standardize_text(series: pd.Series, lowercase: bool = True, 
                    remove_punctuation: bool = False) -> pd.Series:
    """
    Standardize text data: strip whitespace, optionally lowercase and remove punctuation.
    """
    series = series.astype(str).str.strip()
    
    if lowercase:
        series = series.str.lower()
    
    if remove_punctuation:
        series = series.str.replace('[^\w\s]', '', regex=True)
    
    return series


def convert_to_numeric(df: pd.DataFrame, columns: List[str], 
                      errors: str = 'coerce') -> pd.DataFrame:
    """
    Convert specified columns to numeric, handling common issues like
    currency symbols, percentages, and thousands separators.
    
    Args:
        df: Input DataFrame
        columns: Columns to convert
        errors: 'coerce' (invalid -> NaN) or 'raise'
    """
    df = df.copy()
    
    for col in columns:
        # Remove common non-numeric characters
        df[col] = (df[col]
                   .astype(str)
                   .str.replace('$', '')
                   .str.replace(',', '')
                   .str.replace('%', '')
                   .str.strip())
        
        df[col] = pd.to_numeric(df[col], errors=errors)
    
    return df


def deduplicate(df: pd.DataFrame, subset: Optional[List[str]] = None, 
               keep: str = 'first') -> pd.DataFrame:
    """
    Remove duplicate rows with reporting.
    
    Args:
        df: Input DataFrame
        subset: Columns to consider for duplicates. If None, uses all columns
        keep: 'first', 'last', or False (remove all duplicates)
    """
    initial_count = len(df)
    df = df.drop_duplicates(subset=subset, keep=keep)
    removed = initial_count - len(df)
    
    if removed > 0:
        print(f"Removed {removed} duplicate rows ({removed/initial_count*100:.2f}%)")
    
    return df


def fill_missing_values(df: pd.DataFrame, strategy: dict) -> pd.DataFrame:
    """
    Fill missing values using different strategies per column.
    
    Args:
        df: Input DataFrame
        strategy: Dict mapping column names to fill strategies
                  Options: 'mean', 'median', 'mode', 'ffill', 'bfill', or a specific value
    
    Example:
        strategy = {
            'age': 'median',
            'category': 'mode',
            'price': 0,
            'status': 'ffill'
        }
    """
    df = df.copy()
    
    for col, method in strategy.items():
        if col not in df.columns:
            continue
            
        if method == 'mean':
            df[col].fillna(df[col].mean(), inplace=True)
        elif method == 'median':
            df[col].fillna(df[col].median(), inplace=True)
        elif method == 'mode':
            df[col].fillna(df[col].mode()[0] if not df[col].mode().empty else None, inplace=True)
        elif method == 'ffill':
            df[col].fillna(method='ffill', inplace=True)
        elif method == 'bfill':
            df[col].fillna(method='bfill', inplace=True)
        else:
            df[col].fillna(method, inplace=True)
    
    return df


def get_data_quality_report(df: pd.DataFrame) -> pd.DataFrame:
    """
    Generate a comprehensive data quality report.
    """
    report = pd.DataFrame({
        'dtype': df.dtypes,
        'non_null_count': df.count(),
        'null_count': df.isnull().sum(),
        'null_percentage': (df.isnull().sum() / len(df) * 100).round(2),
        'unique_count': df.nunique(),
        'duplicate_count': len(df) - df.nunique()
    })
    
    return report.sort_values('null_percentage', ascending=False)


def quick_data_check(df: pd.DataFrame, n_top_values: int = 5, 
                     show_correlation: bool = True, correlation_threshold: float = 0.7) -> None:
    """
    Comprehensive data quality check with improved diagnostics.
    
    Args:
        df: Input DataFrame
        n_top_values: Number of top values to show for categorical columns
        show_correlation: Whether to show correlation matrix
        correlation_threshold: Threshold for highlighting strong correlations
    """
    print("=" * 60)
    print("QUICK DATA CHECK REPORT")
    print("=" * 60)
    
    # Basic Info
    print("\n=== SHAPE ===")
    print(f"Rows: {df.shape[0]:,} | Columns: {df.shape[1]}")
    print(f"Memory Usage: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
    
    print("\n=== DATA TYPES ===")
    dtype_counts = df.dtypes.value_counts()
    for dtype, count in dtype_counts.items():
        print(f"{dtype}: {count} columns")
    
    print("\n=== ALL COLUMN NAMES ===")
    print(f"Total: {len(df.columns)} columns")
    for i, col in enumerate(df.columns, 1):
        print(f"{i:3d}. {col} ({df[col].dtype})")
    
    print("\n=== FIRST 5 ROWS ===")
    print(df.head())
    
    # Missing Values - Enhanced
    print("\n=== MISSING VALUES ===")
    missing = df.isnull().sum()
    missing_pct = (missing / len(df) * 100).round(2)
    missing_df = pd.DataFrame({
        'Missing': missing,
        'Percent': missing_pct
    })
    missing_df = missing_df[missing_df['Missing'] > 0].sort_values('Missing', ascending=False)
    
    if len(missing_df) > 0:
        print(missing_df)
    else:
        print("No missing values found!")
    
    # Basic Stats
    print("\n=== NUMERIC SUMMARY ===")
    if len(df.select_dtypes(include='number').columns) > 0:
        print(df.describe())
    else:
        print("No numeric columns found.")
    
    # Constant and Near-Constant Columns
    print("\n=== LOW VARIANCE COLUMNS ===")
    constant_cols = [col for col in df.columns if df[col].nunique() == 1]
    near_constant_cols = [col for col in df.columns 
                          if 1 < df[col].nunique() <= 2 and len(df) > 100]
    
    if constant_cols:
        print(f"Constant (single value): {constant_cols}")
    if near_constant_cols:
        print(f"Near-constant (≤2 unique values): {near_constant_cols}")
    if not constant_cols and not near_constant_cols:
        print("None found")
    
    # High Cardinality Check
    print("\n=== HIGH CARDINALITY COLUMNS ===")
    high_card = [(col, df[col].nunique()) for col in df.columns 
                 if df[col].nunique() > len(df) * 0.5 and df[col].nunique() > 50]
    if high_card:
        for col, unique_count in high_card:
            print(f"{col}: {unique_count:,} unique values ({unique_count/len(df)*100:.1f}% of rows)")
    else:
        print("None found")
    
    # Mixed Data Types Check
    print("\n=== MIXED DATA TYPES CHECK ===")
    mixed_found = False
    for column in df.columns:
        inferred_type = pd.api.types.infer_dtype(df[column], skipna=True)
        if inferred_type in ['mixed', 'mixed-integer', 'mixed-integer-float']:
            print(f"⚠️  Column '{column}': {inferred_type}")
            mixed_found = True
    if not mixed_found:
        print("All columns have consistent types ✓")
    
    # Duplicates
    print("\n=== DUPLICATE ROWS ===")
    dup_count = df.duplicated().sum()
    if dup_count > 0:
        print(f"Found {dup_count:,} duplicate rows ({dup_count/len(df)*100:.2f}%)")
    else:
        print("No duplicate rows found ✓")
    
    # Categorical Columns
    print("\n=== TOP VALUES FOR CATEGORICAL COLUMNS ===")
    categorical_cols = df.select_dtypes(include=['object', 'category']).columns
    
    if len(categorical_cols) > 0:
        for col in categorical_cols:
            unique_count = df[col].nunique()
            print(f"\n{col} ({unique_count} unique values):")
            print(df[col].value_counts(dropna=False).head(n_top_values))
    else:
        print("No categorical columns found.")
    
    # Outliers Detection (simple version)
    print("\n=== POTENTIAL OUTLIERS (IQR METHOD) ===")
    numeric_cols = df.select_dtypes(include='number').columns
    outlier_summary = []
    
    for col in numeric_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        outliers = ((df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))).sum()
        if outliers > 0:
            outlier_summary.append({
                'Column': col,
                'Outlier_Count': outliers,
                'Outlier_Pct': f"{outliers/len(df)*100:.2f}%"
            })
    
    if outlier_summary:
        print(pd.DataFrame(outlier_summary))
    else:
        print("No significant outliers detected")
    
    # Correlation Matrix
    if show_correlation and len(numeric_cols) > 1:
        print("\n=== CORRELATION MATRIX (NUMERIC ONLY) ===")
        corr_matrix = df[numeric_cols].corr()
        print(corr_matrix)
        
        # Highlight strong correlations
        print(f"\n=== STRONG CORRELATIONS (|r| > {correlation_threshold}) ===")
        strong_corr = []
        for i in range(len(corr_matrix.columns)):
            for j in range(i+1, len(corr_matrix.columns)):
                corr_val = corr_matrix.iloc[i, j]
                if abs(corr_val) > correlation_threshold:
                    strong_corr.append({
                        'Column 1': corr_matrix.columns[i],
                        'Column 2': corr_matrix.columns[j],
                        'Correlation': f"{corr_val:.3f}"
                    })
        
        if strong_corr:
            print(pd.DataFrame(strong_corr))
        else:
            print(f"No correlations above {correlation_threshold} threshold")
    
    print("\n" + "=" * 60)
    print("END OF REPORT")
    print("=" * 60)
```

```python
train_length = int(len(X)*.8)

X_train = X[:train_length]
X_test = X[train_length:]
y_train = y[:train_length]
y_test = y[train_length:]


print(f"X_train shape: {X_train.shape}")  # Should be (18360, 15, 9)
print(f"y_train shape: {y_train.shape}")  # Should be (18360, 15)
```

```python
lstm_window_size = 3
window = np.lib.stride_tricks.sliding_window_view(x=X_train, window_shape=(lstm_window_size, X_train.shape[1]))
X_train = window.reshape(window.shape[0], lstm_window_size, X_train.shape[1])
y_train = y_train[lstm_window_size-1:]
```

```python
window = np.lib.stride_tricks.sliding_window_view(x=X_test, window_shape=(lstm_window_size, X_test.shape[1]))
X_test = window.reshape(window.shape[0], lstm_window_size, X_test.shape[1])
y_test = y_test[lstm_window_size-1:]
```

```python
X_train
```

# Run LSTM (RNN)

```python
epochs = 30
batch_size = 16
n_hidden = 64                   # Up from 32
timesteps = len(X_train[0])     # Should be 15 (stations)
input_dim = len(X_train[0][0])  # Should be 9 (observation types)
n_classes = len(y_train[0])     # Should be 15 (stations to predict)

model = Sequential()
model.add(LSTM(n_hidden, input_shape=(timesteps, input_dim)))
model.add(Dropout(0.2))         # Down from 0.5
model.add(Dense(n_classes, activation='sigmoid'))

# Verification
print(f"timesteps: {timesteps}")  # Should be 15
print(f"input_dim: {input_dim}")  # Should be 9
print(f"n_classes: {n_classes}")  # Should be 15
```

```python
model.compile(loss='binary_crossentropy', # Changed from categorical_crossentropy based on AI feedback
              optimizer='rmsprop',
              metrics=['binary_accuracy'])
```

```python
model.fit(X_train,
          y_train,
          batch_size=batch_size,
          validation_data=(X_test, y_test),
          epochs=epochs)
```

## Confusion Matrix

```python
# Make predictions on test set
y_pred = model.predict(X_test)

# Convert probabilities to binary predictions (threshold of 0.5)
y_pred_binary = (y_pred > 0.5).astype(int)

# Now create confusion matrix
from sklearn.metrics import classification_report, multilabel_confusion_matrix

print(classification_report(y_test, y_pred_binary))

# Confusion matrix for each of the 15 stations
cm = multilabel_confusion_matrix(y_test, y_pred_binary)
print(f"Confusion matrix shape: {cm.shape}")  # Should be (15, 2, 2)
```
